```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
np.random.seed(1)
x = np.linspace(0, 8, 90)
y1 = 3 + 4*x/8 + np.random.uniform(0.0, 7, len(x))
y2 = 1 + 2*x/8 + np.random.uniform(0.0, 5, len(x))

# plot
fig, ax = plt.subplots()

ax.fill_between(x, y1, y2, alpha=.5, linewidth=2,color="purple",linestyle="solid")
ax.plot(x, (y1 + y2)/2, linewidth=3,color="pink",linestyle="dotted")

ax.set(xlim=(1, 8), xticks=np.arange(1, 8),
       ylim=(0, 12), yticks=np.arange(4, 5))


fig.set_size_inches(18,3.5)
plt.show()
```
