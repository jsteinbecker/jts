```python
# check if emp is trained for sft
canWork = lambda e,s: str(s) in [y for y in str(df['worksShifts'].iloc[e])]
# does emp like this sft
doesPrefer = lambda e,s: str(s) in [y for y in str(df['prefers'].iloc[e])]
# does sft work on this day
```

```python
"""CHECKS IF SHIFT IS SCHEDULED ON A PARTICULAR DAY"""

shiftExists = lambda d,s: str(s) in [y for y in str(df['day'+str(d)].iloc[15])]
shiftExists(5,7)
```

```python
getEmpl2weeks = lambda e: [x for x in df.loc[e,"day1":"day14"]]
getEmpl2weeks(6)
```

```python
getEmp1week = lambda e: [x for x in df.loc[e,"day1":"day7"]]
def w1under (e):
   c = 0
   for x in getEmp1week(e):
      if 0 < x < 10:
         c += 1
   if c < 4:
      return True
   return False

getEmp2week = lambda e: [x for x in df.loc[e,"day8":"day14"]]
def w2under (e):
   c = 0
   for x in getEmp2week(e):
      if 0 < x < 10:
         c += 1
   if c < 4:
      return True
   return False

under40perWeek = lambda e: w2under(e) & w1under(e)
under40perWeek(1)
```

```python
shiftsOnDay = lambda d: [x for x in df.loc[0:14,"day"+str(d)]]
shiftsOnDay(1)
   # STATE:
```

```python
isShiftNeededOnDay = lambda d,s: not s in shiftsOnDay(d)
isShiftNeededOnDay(1,9)
```

```python
def underFTE(e):
   count = 0
   emp2w = [x for x in getEmpl2weeks(e)]
   for x in emp2w:
      if 0 < x < 10:
         count += 1
   if count < df['fte'][e]:
      return True
   return False
underFTE(0)
```

```python
"REQUIRED RULES"
def solve (e,d,s):
   df.iloc[e,d+4] = s
   if canWork(e,s) & shiftExists(d,s) & isShiftNeededOnDay(d,s) & underFTE(e) & under40perWeek(e):
      return True
   df.iloc[e,d+4] = 0
   return False
```
